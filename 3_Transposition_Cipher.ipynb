### Consider the plaintext “DEPARTMENT OF COMPUTER SCIENCE AND TECHNOLY UNIVERSITY OF RAJSHAHI BANGLADESH”, find out the corresponding Transposition Cipher (Take width as input). Then perform the reverse operation to get original plaintext.

## Function for Encryption

```python
def encrypt(plaintext, width = 5):
    length = len(plaintext)
    ciphertext = ""

    for i, char in enumerate(plaintext):
        print(char, end=" ")
        if (i + 1) % width == 0:
            print()
    print()

    for k in range(width):
        for i in range(k, length, width):
            ciphertext += plaintext[i]

    return ciphertext
```

## Function for Decryption

```python
def decrypt(cipher_text, width = 5):
    length = len(cipher_text)
    plain_text = [''] * length
    idx = 0

    for k in range(width):
        for i in range(k, length, width):
            plain_text[i] = cipher_text[idx]
            idx += 1

    return ''.join(plain_text)
```

## Example Usage

```python
plaintext = "DEPARTMENT OF COMPUTER SCIENCE AND TECHNOLY UNIVERSITY OF RAJSHAHI BANGLADESH"
ciphertext = encrypt(plaintext)
decrypted_text = decrypt(ciphertext)
print(f"Plaintext:           {plaintext}")
print(f"Ciphertext:          {ciphertext}")
print(f"Decrypted Plaintext: {decrypted_text}")
```
