### Find out corresponding double Transposition Cipher of the plaintext “DEPARTMENT OF COMPUTER SCIENCE AND TECHNOLY UNIVERSITY OF RAJSHAHI BANGLADESH”. Then perform the reverse operation to get original plaintext.

## Function for Encryption

```python
def encrypt(plaintext, width = 5):
    length = len(plaintext)
    ciphertext = ""

    for i, char in enumerate(plaintext):
        print(char, end=" ")
        if (i + 1) % width == 0:
            print()
    print()

    for k in range(width):
        for i in range(k, length, width):
            ciphertext += plaintext[i]

    return ciphertext
```

## Function for Decryption

```python
def decrypt(cipher_text, width = 5):
    length = len(cipher_text)
    plain_text = [''] * length
    idx = 0

    for k in range(width):
        for i in range(k, length, width):
            plain_text[i] = cipher_text[idx]
            idx += 1

    return ''.join(plain_text)
```

## Example Usage

```python
plaintext = "DEPARTMENT OF COMPUTER SCIENCE AND TECHNOLY UNIVERSITY OF RAJSHAHI BANGLADESH"
print("First Transposition:")
ciphertext1 = encrypt(plaintext)
print("Second Transposition:")
ciphertext2 = encrypt(ciphertext1)
decrypted_text2 = decrypt(ciphertext2)
decrypted_text1 = decrypt(decrypted_text2)
print(f"Plaintext:             {plaintext}")
print(f"Ciphertext1:           {ciphertext1}")
print(f"Ciphertext2:           {ciphertext2}")
print(f"Decrypted Ciphertext1: {decrypted_text2}")
print(f"Decrypted Plaintext:   {decrypted_text1}")
```
